```python
import pandas as pd
```

```python
df1=pd.read_csv('/content/drive/MyDrive/Data for WQU/mexico-real-estate-1.csv')
df2=pd.read_csv('/content/drive/MyDrive/Data for WQU/mexico-real-estate-2.csv')
df3=pd.read_csv('/content/drive/MyDrive/Data for WQU/mexico-real-estate-3.csv')
```

```python
print("type of df1", type(df1))
print("shape of df1", df1.shape)
print("type of df2", type(df2))
print("shape of df2", df2.shape)
print("type of df3", type(df3))
print("shape of df3", df3.shape)
```

```python
# Print df1 shape
df1.shape

# Print df1 info
df1.info()
# Get output of df1 head
df1.head()
```

```python
# Drop null values from df1
df1.dropna(inplace = True)
# Clean "price_usd" column in df1
df1["price_usd"] = df1["price_usd"].str.replace("$","",regex=False)
df1["price_usd"] = df1["price_usd"].str.replace(",","",regex=False).astype(float)

# Print object type, shape, and head
print("df1 type:", type(df1))
print("df1 shape:", df1.shape)
df1.info()
```

```python
df2.head()
```

```python
# Drop null values from df2
df2.dropna(inplace = True)
# Create "price_usd" column for df2 (19 pesos to the dollar in 2014)
df2["price_usd"] = df2["price_mxn"]/19

# Drop "price_mxn" column from df2
df2 = df2.drop("price_mxn", axis=1)
# Print object type, shape, and head
print("df2 type:", type(df2))
print("df2 shape:", df2.shape)
df2.head()
```

```python
df3.head()
```

```python
df3["lat-lon"].str.split(",",expand = True)[0]
```

```python
# Drop null values from df3
df3.dropna(inplace = True)
# Create "lat" and "lon" columns for df3

df3[["lat", "lon"]] = df3["lat-lon"].str.split(",",expand=True)

# Print object type, shape, and head
print("df3 type:", type(df3))
print("df3 shape:", df3.shape)
df3.head()
```

```python
df3["place_with_parent_names"].str.split("|",expand = True)[2].head()
```

```python
# Create "state" column for df3
df3["state"] = df3["place_with_parent_names"].str.split("|",expand = True)[2]

# Drop "place_with_parent_names" and "lat-lon" from df3
df3 = df3.drop(["lat-lon","place_with_parent_names"],axis=1)

# Print object type, shape, and head
print("df3 type:", type(df3))
print("df3 shape:", df3.shape)
df3.head()
```

```python
# Concatenate df1, df2, and df3
df = pd.concat([df1,df2,df3],axis=0)

# Print object type, shape, and head
print("df type:", type(df))
print("df shape:", df.shape)
df.head()
```

```python
#describing the price_usd and the area_m2
df[["area_m2","price_usd"]].describe()
```

```python
#plotting a histogram to understand the skewness of the distribution of the area of the houses
plt.hist(
    df["area_m2"]
)
plt.xlabel("area_m2")
plt.ylabel("frequency")
plt.title("Histogram Showing Distribution of area/m2 for houses in Mexico");
```

```python
#One more way other than a histogram to unstand the skewness is a box plot
plt.boxplot(
    df["area_m2"], vert = False)
plt.xlabel("Area/m2");
plt.title("Distribution of Home sizes in mexico")
```

```python
plt.hist(df["price_usd"])
plt.xlabel("price in US dollars")
plt.ylabel("frequency")
```

```python
plt.boxplot(df["price_usd"], vert=False)
plt.xlabel("price in US dollars")
plt.ylabel("frequency");
#the skew is so extreme that the points outside are outliers
```
